Task - 3 - Outlier Detection:

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.

```python
# Calculate IQR for Score
Q1_score = south_asia_df['score'].quantile(0.25)
Q3_score = south_asia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Define outlier boundaries for Score
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Identify outlier countries based on Score
score_outliers = south_asia_df[(south_asia_df['score'] < lower_bound_score) | (south_asia_df['score'] > upper_bound_score)]

# Calculate IQR for GDP per Capita (Log GDP)
Q1_gdp = south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Define outlier boundaries for GDP per Capita
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outlier countries based on GDP per Capita
gdp_outliers = south_asia_df[(south_asia_df['Log GDP per capita'] < lower_bound_gdp) | (south_asia_df['Log GDP per capita'] > upper_bound_gdp)]

# Combine both outliers
outliers = pd.concat([score_outliers, gdp_outliers]).drop_duplicates()

# Display the outliers
print("Outlier Countries Based on Score and GDP per Capita:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.

```python
import matplotlib.pyplot as plt

# Plot all the data points first
plt.figure(figsize=(10, 6))
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], color='blue', label='Normal Data', alpha=0.7)

# Highlight the outliers in red
plt.scatter(score_outliers['Log GDP per capita'], score_outliers['score'], color='red', label='Score Outliers', alpha=1.0)
plt.scatter(gdp_outliers['Log GDP per capita'], gdp_outliers['score'], color='orange', label='GDP Outliers', alpha=1.0)

# Add title and labels
plt.title('Scatter Plot: GDP per Capita vs Score (Outliers Highlighted)', fontsize=16)
plt.xlabel('Log GDP per capita', fontsize=12)
plt.ylabel('Score', fontsize=12)

# Add legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

6. Outlier Detection:

• Identify outlier countries in both regions based on Score and GDP per Capita.

```python
# Function to detect outliers based on IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Detect outliers for Score in South Asia
south_asia_score_outliers = detect_outliers(south_asia_df, 'score')

# Detect outliers for GDP per Capita in South Asia
south_asia_gdp_outliers = detect_outliers(south_asia_df, 'Log GDP per capita')

# Detect outliers for Score in Middle East
middle_east_score_outliers = detect_outliers(middle_east_df, 'score')

# Detect outliers for GDP per Capita in Middle East
middle_east_gdp_outliers = detect_outliers(middle_east_df, 'Log GDP per capita')

# Display the outliers
print("Outlier Countries in South Asia based on Score:")
print(south_asia_score_outliers[['Country name', 'score', 'Log GDP per capita']])
print("\nOutlier Countries in South Asia based on GDP per Capita:")
print(south_asia_gdp_outliers[['Country name', 'score', 'Log GDP per capita']])

print("\nOutlier Countries in the Middle East based on Score:")
print(middle_east_score_outliers[['Country name', 'score', 'Log GDP per capita']])
print("\nOutlier Countries in the Middle East based on GDP per Capita:")
print(middle_east_gdp_outliers[['Country name', 'score', 'Log GDP per capita']])
```

• Plot these outliers and discuss their implications.

```python
# Plot for South Asia: Score vs GDP per Capita with outliers marked
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, label='Normal', color='blue')
sns.scatterplot(x=south_asia_score_outliers['Log GDP per capita'],
                y=south_asia_score_outliers['score'], color='red', label='Score Outliers', s=100, marker='X')
sns.scatterplot(x=south_asia_gdp_outliers['Log GDP per capita'],
                y=south_asia_gdp_outliers['score'], color='green', label='GDP Outliers', s=100, marker='X')

plt.title("South Asia: Score vs GDP per Capita with Outliers")
plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score (Score)")
plt.legend()
plt.show()

```
